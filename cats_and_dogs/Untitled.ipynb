```python
import tensorflow_datasets as tfds
import tensorflow as tf
import numpy as np
import cv2
import keras
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D, Dense, Flatten
```

```python
from sklearn.model_selection import train_test_split
```

```python
ds,info = tfds.load("CatsVsDogs", split='train', try_gcs=True, with_info=True)
assert isinstance(ds, tf.data.Dataset)
```

```python
len(ds)
```

```python
info
```

```python
images = []
labels = []
j=0
for i in ds:
    if 32<=i['image'].numpy().shape[0]:
        images.append(i['image'].numpy())
        labels.append(i['label'].numpy())
        j+=1
    if j==15000:
        break
```

```python
min_v = images[0].shape[0]
for i in images:
    if min_v>i.shape[0]:
        min_v=i.shape[0]
        over = i.shape
print(min_v)
print(over)
```

```python
small_shape_images = []
for image in images:
    small_shape_images.append(cv2.resize(image, dsize = (32,32), interpolation=cv2.INTER_CUBIC))
```

```python
small_shape_images = np.array(small_shape_images)
labels = np.array(labels)
```

```python
labels
```

```python
x_train,x_test,y_train,y_test = train_test_split(small_shape_images,labels,test_size=0.3, shuffle=True)
```

```python
x_train.shape
```

```python
model = Sequential([
    Conv2D(16, (3,3), padding='same', activation='relu', input_shape=(32,32,3) ),
    MaxPool2D(2,2),
    Conv2D(32, (3,3), padding='same', activation='relu'),
    MaxPool2D(2,2),
    Conv2D(68, (3,3), padding='same', activation='relu'),
    MaxPool2D(2,2),
    Conv2D(128, (3,3), padding='same', activation='relu'),
    
    Flatten(),
    Dense(512, activation='relu'),
    Dense(256, activation='relu'),
    Dense(100, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
opt = keras.optimizers.Adam(learning_rate=0.01)
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# y_train = np.asarray(y_train).astype('float32').reshape((-1,1))
# y_test = np.asarray(y_test).astype('float32').reshape((-1,1))
```

```python
model.fit(x_train, y_train, epochs=20)
```

```python
model.evaluate(x_test,y_test, batch_size=32)
```
